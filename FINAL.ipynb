# Step 2: Create a Simple Indexing System

## Prepare a Dataset

```python
## DONT RERUN THIS CODE

pdf_folder = "/content/drive/MyDrive/IAT 360 Final Project/data/raw"
pdf_files = [os.path.join(pdf_folder, f) for f in os.listdir(pdf_folder) if f.endswith(".pdf")]

!pip install PyPDF2
from PyPDF2 import PdfReader

# Define a small dataset (corpus) of knowledge
documents = []
    # "Retrieval-Augmented Generation (RAG) combines retrieval of relevant documents with text generation to produce more accurate responses.",
    # "RAG uses two main components: a retriever to find relevant information and a generator to create responses based on that information.",
    # "One advantage of RAG is its ability to provide grounded and fact-based answers by retrieving external knowledge.",
    # "The retriever in RAG searches a knowledge base to find documents most similar to the user's query.",
    # "The generator in RAG uses a language model to combine retrieved documents with the query to generate a coherent response.",
    # "RAG systems are used in applications like chatbots, customer support, and question-answering systems.",
    # "A key benefit of RAG is that it reduces hallucination by grounding the responses in retrieved content.",
    # "Dense vector embeddings, created using models like SentenceTransformers, help the retriever find relevant documents in RAG."

for pdf_path in pdf_files:
  reader = PdfReader(pdf_path)
  text = ""

  for page in reader.pages:
    text += page.extract_text() or ""

  documents.append(text)
```

```python
import pickle

path = "/content/drive/MyDrive/IAT 360 Final Project/dataset/raw/documents.pkl"

with open(path, "rb") as f:
    documents = pickle.load(f)

print("Type:", type(documents))
print("Length:", len(documents))
print("Sample:", documents[:1])
```

## Chunk ODP Text

```python
def chunk_text(text, chunk_size=500, overlap=50):
    """
    Splits long text into overlapping chunks.

    Parameters:
        text (str): full document text
        chunk_size (int): number of characters per chunk
        overlap (int): overlapping characters between chunks

    Returns:
        List of text chunks
    """
    chunks = []
    start = 0

    while start < len(text):
        end = start + chunk_size
        chunk = text[start:end]
        chunks.append(chunk)
        start = end - overlap  # overlap to preserve context

    return chunks

# Apply chunking to each document
all_chunks = []
for doc in documents:
    chunks = chunk_text(doc)
    all_chunks.extend(chunks)

print("Total chunks:", len(all_chunks))
print("Sample chunk:", all_chunks[0][:300])
```

## Create Dense Embeddings

```python
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

embeddings = embedding_model.encode(all_chunks)
print("Embedding shape:", embeddings[0].shape)
```

## Simple Searchable Index

```python
df = pd.DataFrame({
    "Chunk": all_chunks,
    "Embedding": list(embeddings)
})

df.head()
```

## Retrieval Function

```python
import numpy as np

def retrieve_with_pandas(query, top_k=3):
    query_embedding = embedding_model.encode([query])[0]

    df["Similarity"] = df["Embedding"].apply(
        lambda emb: np.dot(query_embedding, emb) /
        (np.linalg.norm(query_embedding) * np.linalg.norm(emb))
    )

    return df.sort_values(by="Similarity", ascending=False).head(top_k)
```

## RAG Generation Function

```python
def generate_with_rag(query, top_k=3):
    # retrieve context chunks
    results = retrieve_with_pandas(query, top_k=top_k)
    context = " ".join(results["Chunk"].tolist())

    prompt = f"""
You are a helpful assistant that explains Vancouver's Official Development Plan in simple language.

Context: {context}

Question: {query}

Answer in clear and plain language.
"""

    response = llm(
        prompt,
        max_new_tokens=200,
        num_return_sequences=1
    )[0]["generated_text"]

    return response
```

## Ask Question with RAG

```python
query = "What does zoning mean for me as a student?"
print(generate_with_rag(query))
```

# Compare the Results

```python
query = "What does zoning mean for me as a student?"

# Generate response without RAG
response_no_rag = llm(
    query,
    max_new_tokens=200,  # ensures enough output tokens
    num_return_sequences=1
)[0]['generated_text']

response_rag = generate_with_rag(query, top_k=3)

print("=== Without RAG ===")
print(response_no_rag)
print("\n=== With RAG ===")
print(response_rag)
```
